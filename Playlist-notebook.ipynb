### Loop through the folders and check for playlist file.



```python
import os
import json

from youtube_music_playlist_downloader import (
    get_existing_playlists,
    get_playlist_info,
    format_file_name,
    get_song_file_infos,
    download_song_and_update,
)


folder_name = "~/Music/Playlists"
full_folder = os.path.expanduser(folder_name)


exists = os.path.exists(full_folder)

if exists is False:
    print(f"full path {full_folder}\nDoes not exist\n")
    exit()

full_path = os.walk(full_folder)
print("full path", list(full_path))
os.chdir(full_folder)
```

### Look for existing folder

```python
playlist_data = get_existing_playlists(full_folder, ".playlist_config.json")
file_database = {}

for playlist in playlist_data:
    config = playlist["config_file"]
    f = open(config)
    data = json.load(f)
    playlist_info = get_playlist_info(data)

    entries = playlist_info["entries"]
    playlist_name = playlist_info["title"]
    song_info = get_song_file_infos(format_file_name(playlist_name))

    for video_id in song_info.keys():
        song = song_info[video_id]
        if video_id not in file_database:
            file_database[video_id] = {
                "playlist_name": playlist_name, 
                "file_name": song.file_name,
                "file_path": song.file_path,
                "full_file_path": os.path.join(playlist_name, song.file_name),
            }
```
